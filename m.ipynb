```python
import os
import zipfile

dist_path = "dist"
folder_to_zip = "HabitRecoder.app"
output_zip = "HabitRecoder-Mac.app.zip"

# full paths
folder_path = os.path.join(dist_path, folder_to_zip)
zip_path = os.path.join(dist_path, output_zip)

with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            full_path = os.path.join(root, file)
            relative_path = os.path.relpath(full_path, dist_path)
            zipf.write(full_path, relative_path)

print("Zipped successfully:", zip_path)
```
